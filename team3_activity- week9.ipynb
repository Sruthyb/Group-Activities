3) The manager wants to find out whether the same amount was spent for the three advertisements (TV , Radio and Newspaper). Comment on your findings.

Stating hypothesis:

H0: Amount spent for three ads are equal. 

Ha: Amount spent for three ads are not equal. 

Significance level: 0.05

```python
import scipy
from scipy import stats
from scipy.stats import f_oneway
```

```python
f, p= scipy.stats.f_oneway(data['TV'],data['Radio'], data['Newspaper'])
print(f, p)
if p > 0.05:
    print('Accept Null Hypothesis ')
else :
    print ('Reject Null Hypothesis')
```

Findings:

• f-statistic = 358.85146, pvalue = 4.552931539744962e-103

We are rejecting null hypothesis and accept alternate hypothesis which states that the amount spent for three advertisements are not equal.
