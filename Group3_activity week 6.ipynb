### 2. Examine the data by using visualizations.

### 4.  On average which region has the highest.

####      a. GDP per capita (current US$)

####    b. International trade: Imports (million US$)

```python
# 4(a). sorting values by GDP per capita (current US$)
df[['Region','GDP per capita (current US$)']].groupby('Region').mean().sort_values('GDP per capita (current US$)',ascending=False).head()
```

Findings: 
    
    On average we can see north american region tops on GDP per capita compared to other regions.
    
    

```python
# 4(b). Sorting values by International trade
df[['Region','International trade: Imports (million US$)']].groupby('Region').mean().sort_values('International trade: Imports (million US$)',ascending=False).head()
```

Findings:
    
    On International trade, again north american region tops the list.

### 5.Find out which region has more consistent with respect to GDP percapita (current US$)

```python
# method 1: by calculating mode
df[['Region','GDP per capita (current US$)']].mode()
```

```python
# method 2: By Coefficient of variation
df_new = df.groupby('Region')['GDP per capita (current US$)'].describe()
df_new['CV'] = (df_new['std']/df_new['mean'])*100
df_new.sort_values(by='CV',ascending=True).head()
```

Findings: 
    
  1) from method 1, The west asian region is more consistent with w.r.t GDP per capita.  
  2) from method 2, oceania region have lower coeficient of variation w.r.t GDP per capita.
